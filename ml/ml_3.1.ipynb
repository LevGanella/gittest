```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://github.com/koroteevmv/ML_course/raw/refs/heads/main/ML3.1%20polynomial%20features/data/gen_regression.csv", index_col=0)
data.head()
```

```python
X = pd.DataFrame(data.x)
y = data.y
```

```python
X.shape , y.shape
```

```python
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression().fit(X, y)
```

```python
plt.scatter(X, y)

xx = np.arange(X.x.min(axis=0), X.x.max(axis=0))
yy = linear.predict(pd.DataFrame(xx))
plt.plot(xx, yy, color='red')
```

```python
linear.score(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
transformer = PolynomialFeatures(degree=2, include_bias=False)
transformer.fit(X)
```

```python
X_2 = transformer.transform(X)
X_2.shape
X_2[:5]
```

```python
X[:5]
```

```python
poly_2d = LinearRegression().fit(X_2, y)
poly_2d.score(X_2, y)
```

```python
plt.scatter(X, y)

xx = np.arange(X.x.min(axis=0), X.x.max(axis=0))
xx_2d = transformer.transform(pd.DataFrame(xx))
yy = poly_2d.predict(xx_2d)
plt.plot(xx, yy, color='red')
```

```python
transformer_2 = PolynomialFeatures(degree=10)
transformer_2.fit(X)
X_3 = transformer_2.transform(X)
poly_3d = LinearRegression().fit(X_3, y)
poly_3d.score(X_3, y)
```
